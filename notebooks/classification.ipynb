```python
from sklearn import datasets

from f3dasm.src.space import ContinuousSpace, DiscreteSpace
from f3dasm.src.designofexperiments import DoE
from f3dasm.src.data import Data

import pandas as pd
```

```python
iris = datasets.load_iris()
```

```python
x1 = ContinuousSpace(name="sepal length")
x2 = ContinuousSpace(name="sepal width")
x3 = ContinuousSpace(name="petal length")
x4 = ContinuousSpace(name="petal width")

y = DiscreteSpace(name="target")
```

```python
design = DoE(input_space=[x1, x2, x3, x4], output_space=[y])
```

```python
input = pd.DataFrame(data=iris.data,columns=[("input",name) for name in design.get_continuous_names()])
```

```python
data = Data(doe=design)
```

```python
data.add(data=input)
data.data
```

```python
data.add_output(iris.target, label="target")
```
